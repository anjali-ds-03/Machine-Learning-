 It is a type of process of finding optimal values for hyperparameters.

Hyperparameters which are parameters of a machine learning model that control the learning process and are set before training begins.

 These settings significantly impact the model's performance like traning speed and quality of results.

# Linear Regression :

1. Solver: it is used for optimisation('auto', 'SVD', 'saga', etc.)
2. Fit Intercept: whether to calculate the intercept  or not(True by default)
3. n_jobs = number of CPU  cores to use during the fit process.

#Lasso and Ridge:

1. alpha value: it is regularisation strength.
default value is 1.0 . Higher the alpha values more shrinkage.
for Ridge: (numerical values)
for Lasso:(l1 and l2)
 for elasticnet : same as above

2. solver: it is used for optimisation('auto', 'SVD', 'saga', etc.)

3. fit  intercept: whether to calculate the intercept  or not(True by default)

4. max_iter: maximum number of iterartion for optimization

5.l1_Ratio (elastic_net): mixed ratio between lasso(1.0) and ridge(0.0)
                                                                   
# Logistic Regression:

   1. C: Inverse of regularization strength, smaller values of C means stronger regularization.

   2. Penalty: type of regularization(l1 and l2)
   3. solver: it is used for optimisation('auto', 'SVD', 'saga', etc.)

   #SVM:
   1. Kernel : "linear", 'rbf', 'poly', 'sigmoid'
   2.  C: Inverse of regularization strength, smaller values of C means stronger regularization.
   3. Degree: (only used whhen kernel is poly)= Degree of polynomial equation.

   # Naive Bayes:
    Few Hyperparameters  but depend on the type:
    1. Gaussian: var_smoothing(for numerical stability)
    2. MultinomialNB : alpha (laplace smoothing)

    # KNN:
    1. n_neighbors: number of neighbors.
    2. metric: Distance Measure.

    # Decision tree:
    1. max_depth: max depth of tree
    2. min_samples_split: minimum sample needed to split an internal node.
    3. min_sample_leaf: minimum samples required at a leaf node.
    4. Criterion: 'gini' or 'entropy' or 'logloss'.

    # Random Forest :
    same as Decision Tree
    1. n_estimators: number of trees in forest.
    2. bootstrap: whether bootstrap samples are used.

    # K means Clustering:
    1. n_clusters number of clusters
    2. max_iter: maximum number of iterartion for optimization
    
    #  DBSCAN:
    1. eps: Radii of all point/ max distance for two samples to be considered on the same neighborhood.
    2. min_samples: minimum numbers of points to form a cluster
    3. metric:  Distance Measure.

    # PCA
    1. n_components: number of components to keep
    2. SVD_solver: it is used for optimisation('auto', 'SVD', 'saga', etc.)




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("diabetes.csv")
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
```

```python
x=data.drop('Outcome', axis=1)
y= data['Outcome']
```

```python
x_train, x_test, y_train , y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
scaler=StandardScaler()
scaler.fit(x_train, y_train)
```

```python
scaler= StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
model=LogisticRegression()

params_grid={
    'C':[0.001,0.01,0.1,1,10,100,1000],
    'penalty':['l1','l2'],
    'solver':['liblinear','saga']
}

grid_log=GridSearchCV(model,params_grid,cv=5,scoring='accuracy')
grid_log.fit(x_train,y_train)
print(f"My best parameters are: {grid_log.best_params_}")
```

```python
model= LogisticRegression(C=0.1,penalty='l1', solver ='saga')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
 from sklearn.svm import SVC


svc=SVC()

svc_params={
    'C':[0.001,0.01,0.1,1,10,100,1000],
    'kernel':['linear','rbf','poly','sigmoid'],
    'gamma': ['scale','auto']
}

grid_svc=GridSearchCV(svc,svc_params,cv=10,scoring='accuracy')
grid_svc.fit(x_train,y_train)
print(f"My best parameters are: {grid_svc.best_params_}")
```

```python
svc=SVC(C=0.1, kernel= 'linear', gamma='scale')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
print (classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC= DecisionTreeClassifier()
DTC.fit(x_train, y_train)
```

```python

DTC_params={
    'max_depth':[5,10,15,20,None],
    'min_samples_split':[4,8,12],
    'min_samples_leaf':[2,4,6,8],
    'criterion':['gini', 'entropy']
}


grid_DTC= GridSearchCV(DTC, DTC_params, cv=10, scoring='accuracy')
grid_DTC.fit(x_train, y_train)
print(f"Best Parameters :{ grid_DTC.best_params_}")
```

```python
DTC= DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_leaf=8, min_samples_split=4)
DTC.fit(x_train, y_train)
y_pred=DTC.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC=RandomForestClassifier()
RFC.fit(x_train, y_train)
```

```python
RFC= RandomForestClassifier()

RFC_params={
    'n_estimators':[50,100,150,200],
    'max_depth':[5,10,15,20,None],
    'min_samples_split':[4,8,12],
    'min_samples_leaf':[2,4,6,8],
    'criterion':['gini', 'entropy']
}
grid_RFC= RandomizedSearchCV(RFC, RFC_params, cv=5, scoring='accuracy')
grid_RFC.fit(x_train, y_train)
y_pred=grid_RFC.predict(x_test)

print(f' Best Parameters :{grid_RFC.best_params_}')

```

```python
RFC=RandomForestClassifier(criterion='entropy', max_depth=10, min_samples_leaf=4, min_samples_split=12, n_estimators=150)
RFC.fit(x_train, y_train)
y_pred=RFC.predict(x_test)
```

```python
print(classification_report(y_pred,y_test))
```
